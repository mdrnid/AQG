```python
# from typing_extensions import TypedDict
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage,BaseMessage,  RemoveMessage
# from langgraph.checkpoint.memory import InMemorySaver

import os
from dotenv import load_dotenv
```

```python
# Load environment variables
load_dotenv()

# print("openAi version : ", OpenAI._version_)

llm_openrouter = ChatOpenAI(
    model=os.getenv("OPENROUTER_MODEL"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    seed= 365,
    temperature= 0 ,
    max_completion_tokens= 100
)

# Test simple completion
print("🧪 Testing qwen3.6 connection...")

try:
    message = HumanMessage(content="Please write a short fairy tale. Keep it short.")

    response = llm_openrouter.invoke([message])

    print("📖 Fairy Tale:")
    print(response.content)
    print(f"\n📊 Tokens: {response.response_metadata.get('token_usage', 'N/A')}")

    print(f"\n✅ Koneksi berhasil!")

except Exception as e:
    print(f"\n❌ Error: {e}")
    print("\nTroubleshooting:")
    print("1. Cek API key di file .env")
    print("2. Pastikan internet connection aktif")
    print("3. Verify API key valid di dashboard Z.AI")
```
