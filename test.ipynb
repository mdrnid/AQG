```python
# from typing_extensions import TypedDict
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage,BaseMessage,  RemoveMessage
# from langgraph.checkpoint.memory import InMemorySaver

import os
from dotenv import load_dotenv
```

```python
# Load environment variables
load_dotenv()

# print("openAi version : ", OpenAI._version_)

llm_openrouter = ChatOpenAI(
    model=os.getenv("ZAI_MODEL"),
    api_key=os.getenv("ZAI_API_KEY"),
    base_url="https://api.z.ai/api/paas/v4",
    # seed= 365,
    # temperature= 7 ,
    # max_completion_tokens= 100
)

# Test simple completion
print("🧪 Testing GLMs connection...")

try:
    message = HumanMessage(content="Please write a short fairy tale. Keep it short.")

    response = llm_openrouter.invoke([message])

    print("📖 Fairy Tale:")
    print(response.content)
    print(f"\n📊 Tokens: {response.response_metadata.get('token_usage', 'N/A')}")

    print(f"\n✅ Koneksi berhasil!")

except Exception as e:
    print(f"\n❌ Error: {e}")
    print("\nTroubleshooting:")
    print("1. Cek API key di file .env")
    print("2. Pastikan internet connection aktif")
    print("3. Verify API key valid di dashboard Z.AI")
```

```python
from dotenv import load_dotenv
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage

# Load environment variables
load_dotenv()

# Inisialisasi model Gemini (rekomendasi resmi)
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",           # atau "gemini-1.5-pro", "gemini-1.5-flash", dll.
    google_api_key=os.getenv("GOOGLE_API_KEY"),
    temperature=0,
    # max_tokens=100,
    # top_p=0.95,                           # parameter Gemini-specific (opsional)
    # safety_settings="BLOCK_MEDIUM_AND_ABOVE"  # contoh pengaturan safety
)

# Test koneksi
print("🧪 Testing Gemini connection via langchain-google-genai...")

try:
    message = HumanMessage(content="Please write a short fairy tale. Keep it short.")
    response = llm.invoke([message])
    
    print("📖 Fairy Tale:")
    print(response.content)
    print(f"\n📊 Tokens: {response.response_metadata.get('token_usage', 'N/A')}")
    print("\n✅ Koneksi berhasil! Menggunakan integrasi resmi LangChain.")
    
except Exception as e:
    print(f"\n❌ Error: {e}")
```
